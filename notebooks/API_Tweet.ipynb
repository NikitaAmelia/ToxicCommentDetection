```python

import pandas as pd
import re, requests

# === 1. Baca dataset ===
df = pd.read_csv("hate_speech_ariel.csv")
if "error" in df.columns:
    df = df[df["error"].isna()]
df = df.dropna(subset=["full_text"])
df = df[["full_text"]]

# === 2. Preprocessing ===
def clean_input(text, max_len=300):
    text = str(text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@[A-Za-z0-9_]+", "", text)
    text = re.sub(r"[^A-Za-z0-9\s]+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text[:max_len]

df["clean_text"] = df["full_text"].apply(clean_input)

# === 3. Ambil sample ===
sample_df = df.sample(n=, random_state=42)   # ambil 10 dulu biar cepat

# === 4. API URL (copy dari print backend) ===
url = "https://7df1375e8405.ngrok-free.app/predict"   # ganti dengan URL dari backend

# === 5. Test API ===
for i, comment in enumerate(sample_df["clean_text"], start=1):
    try:
        res = requests.post(url, json={"comment": comment}, timeout=120)
        if res.status_code == 200:
            data = res.json()
            print(f"\n=== Sample {i} ===")
            print("Comment:", comment)
            print("Detected Label:", data.get("detected_label"))
            print("Bot Response:", data.get("bot_response"))
        else:
            print(f"\n=== Sample {i} ERROR ===")
            print("Status:", res.status_code)
            print(res.text)
    except Exception as e:
        print(f"\n=== Sample {i} ERROR ===")
        print("Exception:", str(e))
```
